```python
# using python
# i like python
# * l*k*pyth*n
t=("i like python")
i=0
result=""
while i <len(t): 
    if t[i] in "aeiouAEIOU":
        result += "*"
    else:
        result += t[i]
    i += 1

print(result)
```

```python
s=('enter a sentence')
v='aeiouAEIOU'
res=''
for i in s:
    if i in v:
        res+='*'
    else:
        res+=i
print(res)
```

```python
# check string is palindrome or not
s = input("Enter a string: ")
rev = ""
i = len(s) - 1
while i >= 0:
    rev = rev + s[i]
    i = i - 1
if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```
